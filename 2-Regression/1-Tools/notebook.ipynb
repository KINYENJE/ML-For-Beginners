# Diabetes Prediction

```python
from sklearn import datasets

#load the diabetes dataset
x, y = datasets.load_diabetes(return_X_y=True) #x is the input data, y is the target data (the data we want to predict)

# Print the shape of the data and the first row
print(x.shape) # (442, 10) 442 rows and 10 columns
print(x[0]) # this is the first row of the data
```

### Extract the bmi data from the dataset and create a 2D array of the data.

```python
bmi_data = x[:, 2]
print(bmi_data.shape) # (442,) 442 rows and 1 column

#Reshape to 2D array
bmi_data_2d = bmi_data.reshape(-1, 1) # -1 means that it will infer the number of rows from the number of columns
print(bmi_data_2d[0:10]) # print the first 10 rows 
```

### Split the data into training data and testing data.

```python
from sklearn import model_selection

# Split the data into training and test sets
bmi_train, bmi_test, y_train, y_test = model_selection.train_test_split(bmi_data_2d, y, test_size=0.33) # 33% of the data will be used for testing and 67% for training
```

### Train a linear regression model on the training data.

```python
# Create a linear regression model and train it using the training data
from sklearn import linear_model

model = linear_model.LinearRegression()
model.fit(bmi_train, y_train)
```

### Make prdictions on the testing data. As we have now trained the model.

```python
# Predict the target values using the test data
y_pred = model.predict(bmi_test) # this will predict the target values using the test data

```

### We use matplotlib to plot the data and the regression line.Inorder to figure out how good our model is.

```python
# Create a scatter plot of the test data and the predicted data
import matplotlib.pyplot as plt

# Plot the test data and the predicted data
plt.scatter(bmi_test, y_test, color='black') # plot the test data

# Add a line showing the predicted data
plt.plot(bmi_test, y_pred, color='blue', linewidth=3) # plot the predicted data

#Add labels and a title
plt.xlabel("Scaled BMI")
plt.ylabel("Diabetes Progression")
plt.title("A graph to show the relationship between BMI and Diabetes Progression")

# Draw the plot
plt.show() # this will display the plot

```
